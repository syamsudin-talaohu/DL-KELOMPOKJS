```python
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertForMaskedLM, BertModel
from bert_score import BERTScorer
from rouge import Rouge
from rouge_score import rouge_scorer
```

```python
from bert_score import score

# Define your texts
original_text = "tugas akhir saya mempengaruhi kesehatan mental saya, apa yang harus saya lakukan?"
generated_text = "Jika tugas akhir menjadi sebuah penyebaran stress bagi Anda, maka Anda dapat mengurangi penyebaran stress tersebut dengan beberapa hal sebagai berikut: 1. Pemahaman dan Pengantinan: Prinsip utama dari menghadapi tugas akhir adalah memahami dan mengantininya."

# Calculate BERTScore
P, R, F1 = score([original_text], [generated_text], lang="id", verbose=True)
print(f"BERTScore Precision: {P.mean():.4f}, Recall: {R.mean():.4f}, F1: {F1.mean():.4f}")
```

```python
from rouge_score import rouge_scorer

# Define your texts
original_text = "tugas akhir saya mempengaruhi kesehatan mental saya, apa yang harus saya lakukan?"
generated_text = "Jika tugas akhir menjadi sebuah penyebaran stress bagi Anda, maka Anda dapat mengurangi penyebaran stress tersebut dengan beberapa hal sebagai berikut: 1. Pemahaman dan Pengantinan: Prinsip utama dari menghadapi tugas akhir adalah memahami dan mengantininya."

# ROUGE calculation
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = scorer.score(original_text, generated_text)
print(f"ROUGE-1 Precision: {scores['rouge1'].precision:.4f}, Recall: {scores['rouge1'].recall:.4f}, F1: {scores['rouge1'].fmeasure:.4f}")
print(f"ROUGE-2 Precision: {scores['rouge2'].precision:.4f}, Recall: {scores['rouge2'].recall:.4f}, F1: {scores['rouge2'].fmeasure:.4f}")
print(f"ROUGE-L Precision: {scores['rougeL'].precision:.4f}, Recall: {scores['rougeL'].recall:.4f}, F1: {scores['rougeL'].fmeasure:.4f}")
```
